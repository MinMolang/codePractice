## 3번
(20 점)

제조업체 (manufacturer) 중에서 "chevrolet", "ford", "honda" 의 모든 자동차의 고속도로 연비 (hwy) 의 평균을 구해보세요.

```
query = 'SELECT manufacturer as "제조업체", AVG(hwy) as "평균연비" from mpg group by manufacturer having manufacturer in ( "chevrolet", "ford", "honda" )'

pd.read_sql_query(query, engine)

# 출력 예
```

## 4번
(20 점)

도시 연비 (cty) 가 18 이상은 별도 추가 컬럼에 "pass" 를 18 미만은 "fail" 라는 값이 출력되게 해주세요.

출력할 컬럼은 model, cty 와 조건에 따라 pass 또는 fail 이 있는 컬럼을 출력하면 됩니다.

```
query = 'SELECT model, cty, CASE WHEN cty >= 18 THEN "pass" ELSE "fail" END AS "합격여부" FROM mpg'

pd.read_sql_query(query, engine)

# 출력 예
```

현재 데이터에서 제조회사를 출력하고, 차량(모델) 대수를 출력합니다.

그리고 제조회사는 ABC 순으로, 차량대수는 높은 순서대로 정렬해주세요.

```
query = 'SELECT MANUFACTURER AS "제조회사", MODEL AS "차량모델", COUNT(MODEL) AS "차량(모델) 대수"  from mpg GROUP BY MANUFACTURER, MODEL ORDER BY MANUFACTURER, COUNT(MODEL) DESC'

pd.read_sql_query(query, engine)

# 출력 예
```
